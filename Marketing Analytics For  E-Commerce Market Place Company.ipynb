### 1. Perform Detailed exploratory analysis
a. Define & calculate high level metrics like (Total Revenue, Total quantity, Total products, Total categories, Total sellers, Total locations, Total channels, Total payment methods etc…) 

```python
# calculating total_revenue
total_revenue = order_payments['payment_value'].sum()
print('Total_revenue : ', total_revenue)

# calculating total_quantity
total_quantity = order_items['order_item_id'].sum()
print('Total_quantity : ', total_quantity)

# calculating total_products
total_products = len(products)
print('Total_products : ', total_products)

# calculating total_categories
total_categories = products['product_category_name'].nunique()
print('Total_categories : ', total_categories)

# calculating total_sellers
total_sellers = len(sellers)
print('Total_sellers : ', total_sellers)

# calculating total_locations
total_locations = geo_location['geolocation_state'].nunique()
print('Total_locations : ', total_locations)

# calculating total_channels
total_channels = orders['order_status'].nunique()
print('Total_channels : ', total_channels)

# calculating total_payment_methods
total_payment_methods = order_payments['payment_type'].nunique()
print('Total_payment_methods : ', total_payment_methods)
```

b. Understanding how many new customers acquired every month

```python
# Converting the order_purchase_timestamp column to datetime format
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Extracting the year and month from the order_purchase_timestamp column
orders['order_month_year'] = orders['order_purchase_timestamp'].dt.to_period('M')

# Grouping the orders by month and counting the unique customers
new_customers = orders.groupby('order_month_year')['customer_id'].nunique()

# Plotting the graph
plt.figure(figsize = (6, 4))
plt.plot('new_customers')
plt.title('New Customers Acquired Each Month')
plt.xlabel('Months')
plt.ylabel('New Customers')
plt.show()
```

c. Understand the retention of customers on a month-on-month basis

```python
#Converting the string to datetime format
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])

#Creating a new column for order_month
orders["order_month"] = orders["order_purchase_timestamp"].dt.to_period('M')

#Grouping the data by customer_id and order_month and calculating the number of orders
customer_orders = orders.groupby(["customer_id", "order_month"]).agg({"order_id": "nunique"}).reset_index()

#Creating a pivot table for the above data
customer_orders_pivot = customer_orders.pivot(index="customer_id", columns="order_month", values="order_id")

#Filling the missing values with 0 and converting the object data type to float data type
customer_orders_pivot = customer_orders_pivot.fillna(0).astype(float)
```

d. How the revenues from existing/new customers on month on month basis

```python
# merge orders with order_payments to get payment amount
order_payments.rename(columns = {'payment_value': 'payment_amount'}, inplace=True)
orders = pd.merge(orders, order_payments, on = 'order_id', how = 'left')

# classify customers as existing or new
first_orders = orders.groupby('customer_id')['order_purchase_timestamp'].min().reset_index()
first_orders.columns = ['customer_id', 'first_order_date']
orders = pd.merge(orders, first_orders, on = 'customer_id', how = 'left')
orders['customer_type'] = np.where(orders['order_purchase_timestamp'] == orders['first_order_date'], 'New', 'Existing')

# group orders by customer, month and customer type and calculate revenue
monthly_revenue = orders.groupby(['customer_type', 'order_month'])['payment_amount'].sum().reset_index()

# pivot table to get revenue by month and customer type
monthly_revenue_pivot = pd.pivot_table(monthly_revenue, index = 'order_month', columns = 'customer_type', values = 'payment_amount', fill_value = 0)

# plot the trend
plt.figure(figsize = (10, 6))
monthly_revenue_pivot.plot(kind = 'bar', stacked = True)
plt.title('Monthly Revenue by Customer Type')
plt.xlabel('Month')
```

f. Popular Products by month, seller, state, category.

```python
# merge Order_Items and Products tables
product_details = pd.merge(order_items, products, on = 'product_id')

# group the data by month, seller, state, and category and calculate the total quantity sold for each product
popular_products = product_details.groupby(['seller_id', 'product_category_name', 'product_id'])[['order_item_id']].sum().reset_index()
```

g. Popular categories by state, month

```python
# Merge the tables on order_id, product_id, seller_id
order_items = pd.merge(orders, order_items, on='order_id')
order_items = pd.merge(order_items, products, on='product_id')
order_items = pd.merge(order_items, sellers, on='seller_id')
```

```python
# Create a pivot table to get the count of products sold by category and month
pivot_table = pd.pivot_table(order_items, values='product_id', index=['product_category_name', 'seller_state', pd.Grouper(key='order_purchase_timestamp', freq='M')], aggfunc='count')
```

```python
# Sort the pivot table in descending order
popular_categories = pivot_table.sort_values(by='product_id', ascending=False)
```

```python
# Print the top 10 popular categories by state and month
for state in popular_categories.index.get_level_values(1).unique():
    for month in popular_categories.index.get_level_values(2).unique():
        print('State:', state, 'Month:', month)
        print(popular_categories.loc[(slice(None), state, month), :].head(10))
```

h. List top 10 most expensive products sorted by price

```python
# Sort the Products table by price in descending order
expensive_products = order_items.sort_values(by='price', ascending=False)

# Print the top 10 most expensive products
print(expensive_products[['product_category_name', 'price']].head(10))
```

### 2. Performing Customers/sellers Segmentation

a. Divide the customers into groups based on the revenue generated

b. Divide the sellers into groups based on the revenue generated 

```python
# Calculate revenue generated by each customer
customer_revenue = order_items.groupby('customer_id')['payment_amount'].sum().reset_index()
customer_revenue.columns = ['customer_id', 'revenue']

# Divide customers into groups based on revenue generated
customer_revenue['segment'] = pd.qcut(customer_revenue['revenue'], q=4, labels=['Bronze', 'Silver', 'Gold', 'Platinum'])

# Calculate revenue generated by each seller
seller_revenue = order_items.groupby('seller_id')['payment_amount'].sum().reset_index()
seller_revenue.columns = ['seller_id', 'revenue']

# Divide sellers into groups based on revenue generated
seller_revenue['segment'] = pd.qcut(seller_revenue['revenue'], q=4, labels=['Bronze', 'Silver', 'Gold', 'Platinum'])

print("Customer Segmentation based on Revenue Generated:")
print(customer_revenue)

print("Seller Segmentation based on Revenue Generated:")
print(seller_revenue)
```

### 3. Cross-Selling (Which products are selling together)
Hint: We need to find which of the top 10 combinations of products are selling together in each transaction. (combination of 2 or 3 buying together)

```python
# create a new column 'product_combo' in Order_Items table
order_items['product_combo'] = order_items.groupby('order_id')['product_id'].transform(lambda x: ','.join(x))

# identify top 10 product combinations
top_product_combinations = order_items['product_combo'].value_counts().head(10)

# display top 10 product combinations
print("Top 10 product combinations:\n")
for i, combo in enumerate(top_product_combinations.index):
    print(f"{i+1}. {combo} - {top_product_combinations[combo]} transactions\n")
```

### 4. Payment Behaviour

a. How customers are paying?

```python
# Group the payment data by payment_method and count the number of payments for each method
payment_method_counts = order_payments.groupby('payment_type').size().reset_index(name='counts')

# Print the results
print(payment_method_counts)
```

b. Which payment channels are used by most customers?

```python
# Sort the payment methods by the number of payments in descending order
payment_method_counts = payment_method_counts.sort_values('counts', ascending=False)

# Print the payment method with the most payments
print("The payment method used most is:", payment_method_counts.iloc[0]['payment_type'])
```

### 5. Customer satisfaction towards category & product

a. Which categories (top 10) are maximum rated & minimum rated?

```python
# Join Order_Items and Order_Review_Ratings tables on order_item_id
merged_df = pd.merge(order_items, order_review_ratings, on='order_id', how='left')

# Group the merged data by category and calculate the average rating for each category
category_ratings = merged_df.groupby('product_category_name').agg({'review_score': 'mean'}).reset_index()

# Sort the categories by the average rating in descending order to get the top 10 categories with maximum rating
top10_max = category_ratings.sort_values('review_score', ascending=False).head(10)

# Sort the categories by the average rating in ascending order to get the top 10 categories with minimum rating
top10_min = category_ratings.sort_values('review_score', ascending=True).head(10)

# Print the top 10 categories with maximum rating and minimum rating
print("Top 10 categories with maximum rating:")
print(top10_max)
print("Top 10 categories with minimum rating:")
print(top10_min)
```

b. Which products (top10) are maximum rated & minimum rated?

```python
# Group the merged data by product_id and calculate the average rating for each product
product_ratings = merged_df.groupby('product_id').agg({'review_score': 'mean'}).reset_index()

# Sort the products by the average rating in descending order to get the top 10 products with maximum rating
top10_max = product_ratings.sort_values('review_score', ascending=False).head(10)

# Sort the products by the average rating in ascending order to get the top 10 products with minimum rating
top10_min = product_ratings.sort_values('review_score', ascending=True).head(10)

# Print the top 10 products with maximum rating and minimum rating
print("Top 10 products with maximum rating:")
print(top10_max)
print("Top 10 products with minimum rating:")
print(top10_min)
```

c. Average rating by location, seller, product, category, month etc.

```python
# Group by location, seller, product, category, and month and calculate the average rating
location_ratings = merged_df.groupby(['seller_state'])['review_score'].mean()
seller_ratings = merged_df.groupby(['seller_id'])['review_score'].mean()
product_ratings = merged_df.groupby(['product_id'])['review_score'].mean()
category_ratings = merged_df.groupby(['product_category_name'])['review_score'].mean()
month_ratings = merged_df.groupby(pd.Grouper(key='order_purchase_timestamp', freq='M'))['review_score'].mean()

# Print the average ratings by location, seller, product, category, and month
print("\nAverage rating by location:")
print(location_ratings)
print("\nAverage rating by seller:")
print(seller_ratings)
print("\nAverage rating by product:")
print(product_ratings)
print("\nAverage rating by category:")
print(category_ratings)
print("\nAverage rating by month:")
print(month_ratings)
```
